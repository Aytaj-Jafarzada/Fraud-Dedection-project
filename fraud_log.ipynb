```python
import pandas as pd
import joblib

# Modeli yüklə
model = joblib.load("rf_model8_encoded.pkl")

# Bütün dataseti yüklə
df = pd.read_csv(r'C:/Users/user/Documents/fraud_transactions7.csv')

# Timestamp sütununu datetime formatına çevir
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Yeni sütunlar əlavə et
df['hour'] = df['timestamp'].dt.hour
df['dayofweek'] = df['timestamp'].dt.dayofweek
df['month'] = df['timestamp'].dt.month_name()

# Modelin istədiyi sütunlar
features = ['amount', 'merchant', 'transaction_type', 'location', 'gender',
            'account_balance', 'hour', 'dayofweek', 'transaction_category']

X = df[features]

# Proqnoz ehtimalı
df['fraud_probability'] = model.predict_proba(X)[:, 1]

# Yeni fayla yaz (tam dataset)
df.to_csv(r"C:/Users/user/Documents/fraud_predictions_full10.csv", index=False)
```
